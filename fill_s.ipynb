```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("D:\DBDA\Projects\project 1\combined_al.csv")
#df.head()
```

```python
df.replace('Not Mentioned', np.nan, inplace=True)
```

```python
df1 = df[["Min_exp","Max_exp","Max_sal"]]
```

```python
test_data = df1[df1["Max_sal"].isnull()]
```

```python
df1.dropna(inplace=True)
#df1
```

```python
y_train = df1["Max_sal"]
#y_train
x_train = df1.drop("Max_sal",axis=1)
#x_train
```

```python
x_test = test_data.drop("Max_sal",axis=1)
x_test.isnull().sum()
#x_test
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
hgb = HistGradientBoostingRegressor()
hgb.fit(x_train,y_train)
```

```python
y_pred = hgb.predict(x_test)
y_pred
```

```python
test_data.loc[test_data.Max_sal.isnull(),"Max_sal"] = y_pred
#test_data
```

```python
frames = [df1,test_data]
final_max_sal = pd.concat(frames)
#final_max_sal
final_max_sal.isnull().sum()
```

```python
df2 = df[["Min_exp","Max_exp","Min_sal"]]
test_data1 = df2[df2["Min_sal"].isnull()]
df2.dropna(inplace=True)
df2.isnull().sum()
y_train = df2["Min_sal"]
x_train = df2.drop("Min_sal",axis=1)
x_train
from sklearn.ensemble import HistGradientBoostingRegressor
hgb = HistGradientBoostingRegressor()
hgb.fit(x_train,y_train)
x_test = test_data1.drop("Min_sal",axis=1)
x_test.isnull().sum()
y_pred = hgb.predict(x_test)
test_data1.loc[test_data1.Min_sal.isnull(),"Min_sal"] = y_pred
test_data1
```

```python
frames = [df2,test_data1]
final_Min_sal = pd.concat(frames)
```

```python
df.Max_sal = final_max_sal.Max_sal
df.Min_sal = final_Min_sal.Min_sal
```

```python
df
```

```python
columns_to_round = ['Min_sal', 'Max_sal']
df[columns_to_round] = df[columns_to_round].round()
df
```

```python
df.to_csv(r"D:\DBDA\Projects\project 1\combined_all.csv",index=False)
```
