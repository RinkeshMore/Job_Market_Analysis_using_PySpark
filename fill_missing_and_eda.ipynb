```python
def outliers(datacolumn):
    sorted(datacolumn)
    Q1,Q3 = np.percentile(datacolumn , [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
```

```python
sns.boxplot(x=df["Min_exp"], orient="h")
```

```python
lower_limit_Min_exp,upper_limit_Min_exp = outliers(df.Min_exp)
print(lower_limit_Min_exp,upper_limit_Min_exp)
```
